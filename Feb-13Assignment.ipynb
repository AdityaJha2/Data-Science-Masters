```python
#Q4-Why LookupError class is used? Explain with an example KeyError and IndexError.

print('''The LookupError exception in Python forms the base class for all exceptions that are raised
when an index or a key is not found for a sequence or dictionary respectively.\n''')

print('KeyError-A Python KeyError exception is raised when we try to access a key that isn’t in a dictionary.\nfor example-')
try:
    d={1:[2,3,4],'key':'Aditya'}
    d["k"]
except KeyError as e:
    print(e)
    
print('''\nIndexError-This error occurs when we are try to access an index item in a Python list that is out of range,
meaning that an item doesn’t have an index position.\nfor example-''')
try:
    l=[1,2,3,4,5]
    l[10]
except IndexError as e:
    print(e)
```
