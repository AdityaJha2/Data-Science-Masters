```python
# 2.What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
```

```python
print('''Scatter plot - A scatter plot is a diagram where each value in the data set is represented by a dot. 
We can change the shape of scatter plot using markers .''')
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y, c='#874CCC', alpha=.9)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot of x V/S y')
```

```python
# 3.Why is the subplot() function used? Draw four line plots using the subplot() function.
```

```python
print('''Subplots() function in Python simplifies the creation of multiple subplots Matplotlib within a single figure,
allowing for organized and simultaneous visualization of various datasets or plots''')
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([50, 20, 40, 20, 60, 70])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([10, 20, 30, 40, 50, 60])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([200, 350, 250, 550, 450, 150])
fig = plt.figure()
ax = fig.subplots(nrows=2, ncols=2)
ax[0, 0].plot(x, y, c='red')
ax[0, 1].plot(x1, y1, c='green')
ax[1, 0].plot(x2, y2, c='blue')
ax[1, 1].plot(x3, y3, c='yellow')
plt.show()
```

```python
# 4.What is a bar plot? Why is it used? 
print('''A barplot is one of the most common types of graphic.
It shows the relationship between a numeric and a categoric variable.
Each entity of the categoric variable is represented as a bar''')
```

```python
plt.figure(figsize=(5,3))
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit, color='#4793AF')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company Wise Profit')
```

```python
# 5.What is a box plot? Why is it used?
```

```python
print('''Box plots are used to show distributions of numeric data values, 
especially when you want to compare them between multiple groups. 
They are built to provide high-level information at a glance, offering general information about a
group of data's symmetry, skew, variance, and outliers''')
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
fig = plt.figure(figsize=(4,3))
ax = fig.add_axes([0,0,1,1])
ax.boxplot(data)
plt.show()
```
